```python
import pandas as pd
df = pd.read_csv('fatal-police-shootings-data.csv')
df
```

# Zestawienie jednoczeÅ›nie liczby ofiar interwencji wedÅ‚ug rasy oraz tego, czy wykazywaÅ‚y one oznaki choroby psychicznej.

```python
import numpy as np
df_race_illness = df.pivot_table(values='id',index='race',columns='signs_of_mental_illness',aggfunc=lambda x: len(x.unique()))
df_race_illness
```

# Dodaj do tego zestawienia kolumnÄ™ wskazujÄ…cÄ… jaki odsetek ofiar interwencji wykazywaÅ‚o oznaki choroby psychicznej dla kaÅ¼dej z ras.

```python
def illness_perc(row):
    perc = (row[True]/(row[True] + row[False]))*100
    return perc
```

```python
df_race_illness['illness_perc'] = df_race_illness.apply(lambda row: illness_perc(row),axis=1).round(2)
```

```python
df_race_illness
```

# Odpowiedz, ktÃ³ra z nich charakteryzuje siÄ™ najwiÄ™kszym odsetkiem znamion choroby psychicznej podczas interwencji.

```python
df_race_illness.sort_values(by = ['illness_perc'], ascending=False).iloc[0]
```

# Dodaj kolumnÄ™ oznaczajÄ…cÄ… dzieÅ„ tygodnia, w ktÃ³rym doszÅ‚o do interwencji. Zlicz interwencje wedÅ‚ug odpowiedniego dnia tygodnia. NastÄ™pnie stwÃ³rz wykres kolumnowy, tak aby dni tygodnia byÅ‚y uszeregowane od poniedziaÅ‚ku do niedzieli.

```python
df['day'] = (pd.to_datetime(df['date'])).dt.day_name()
```

```python
order = [ 'Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
df['day'] = pd.Categorical(df['day'], categories=order, ordered=True)
```

```python
df_day = df.pivot_table(values='id',index='day',aggfunc=lambda x: len(x.unique()))
```

```python
df_day
```

```python
import matplotlib.pyplot as plt
plt.bar(df_day.index, df_day['id'])
```

#  do ilu incydentÃ³w w bazie dotyczÄ…cej Å›miertelnych interwencji doszÅ‚o w przeliczeniu na 1000 mieszkaÅ„cÃ³w kaÅ¼dego ze stanÃ³w

```python
populaton = pd.read_html('https://simple.wikipedia.org/wiki/List_of_U.S._states_by_population', header=0)
states = pd.read_html('https://secure.ssa.gov/poms.nsf/lnx/0901501010', header=0)
df_population = populaton[0]
df_states = states[0]
```

```python
df_states_rename = df_states.rename(columns={'Territory Abbreviation':'state'})
```

```python
df_pop = df_population[['State','Population estimate, July 1, 2019[2]']]
```

```python
df_merged_step_1 = pd.merge(df,df_states_rename,how='inner',on='state')
```

```python
df_final_merge = pd.merge(df_merged_step_1,df_pop,how='inner',on='State')
```

```python
df_kills = df_final_merge.pivot_table(values='id',index='State',aggfunc=lambda x: len(x.unique()))
```

```python
df_no_dupl = df_final_merge[['State','Population estimate, July 1, 2019[2]']].drop_duplicates()
```

```python
df_kills_1000 = pd.merge(df_kills,df_no_dupl,how='inner',on='State')
```

```python
df_kills_1000['per_1000'] = (df_kills_1000['id']/df_kills_1000['Population estimate, July 1, 2019[2]'])*1000
```
